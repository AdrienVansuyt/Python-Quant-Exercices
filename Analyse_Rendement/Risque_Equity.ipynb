# **Analyse des Rendements et Risques des Actions Tech**
## **Introduction:**

Dans cet exercice, nous allons analyser les rendements et les risques (volatilité) de quelques grandes actions technologiques (Apple, Microsoft, Google, Amazon, et Tesla) sur une période d'un an. En comparant les rendements moyens et la volatilité de ces actions, nous pouvons obtenir des informations précieuses pour évaluer leurs performances et risques relatifs.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Liste des symboles des actions technologiques
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']

# Télécharger les données de prix
data = yf.download(tickers, period='1y', interval='1d')

# Calculer les rendements quotidiens
returns = data['Close'].pct_change().dropna()

# Calculer les rendements moyens annuels et la volatilité annuelle
mean_returns = returns.mean() * 252
volatility = returns.std() * (252 ** 0.5)

# Créer un DataFrame pour les rendements et la volatilité
risk_return_df = pd.DataFrame({'Mean Return': mean_returns, 'Volatility': volatility})

# Tracer les rendements moyens et la volatilité
plt.figure(figsize=(10, 6))
plt.scatter(risk_return_df['Volatility'], risk_return_df['Mean Return'], c='blue', marker='o')
for i in risk_return_df.index:
    plt.annotate(i, (risk_return_df['Volatility'][i], risk_return_df['Mean Return'][i]))
plt.xlabel('Volatility (Annualized)')
plt.ylabel('Mean Return (Annualized)')
plt.title('Risk vs Return for Major Tech Stocks')
plt.grid(True)
plt.show()
```

## **Conclusion:**

En examinant ce graphique, nous pouvons observer la relation entre le risque et le rendement pour différentes actions technologiques. Cela peut aider les investisseurs à choisir les actions en fonction de leur tolérance au risque et de leurs objectifs de rendement.
