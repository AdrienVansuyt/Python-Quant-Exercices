# **Analyse des Gaps de Prix avec Volume pour Nvidia, le Pétrole et l'Or**
## **Introduction:**

Dans cet exercice, nous allons analyser les gaps de prix à l'ouverture et les volumes associés pour Nvidia, le pétrole, et l'or pour l'année 2024. Les gaps de prix peuvent indiquer une forte pression d'achat ou de vente, et en les associant aux volumes, nous pouvons évaluer la force des mouvements de marché.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Liste des symboles
tickers = ['NVDA', 'CL=F', 'GC=F']

# Télécharger les données pour 2024
data = {ticker: yf.download(ticker, start='2024-01-01', end='2024-12-31') for ticker in tickers}

# Fonction pour détecter les gaps de prix et les volumes associés
def detect_gaps(data, ticker):
    data['Gap'] = data['Open'] - data['Close'].shift(1)
    data['Gap_Percent'] = (data['Gap'] / data['Close'].shift(1)) * 100
    data['Volume'] = data['Volume']
    return data[['Open', 'Close', 'Gap', 'Gap_Percent', 'Volume']]

# Détecter les gaps pour chaque actif
gaps = {ticker: detect_gaps(data[ticker], ticker) for ticker in tickers}

# Créer des graphiques pour visualiser les gaps et les volumes
for ticker in tickers:
    plt.figure(figsize=(14, 7))
    plt.bar(gaps[ticker].index, gaps[ticker]['Gap_Percent'], color='blue', label='Gap Percent')
    plt.xlabel('Date')
    plt.ylabel('Gap Percent')
    plt.twinx()
    plt.plot(gaps[ticker].index, gaps[ticker]['Volume'], color='red', label='Volume')
    plt.ylabel('Volume')
    plt.title(f'{ticker} - Price Gaps and Volume in 2024')
    plt.legend()
    plt.show()
```

# **Conclusion:**

En observant ces graphiques, nous pouvons voir les gaps de prix à l'ouverture pour Nvidia, le pétrole et l'or en 2024, ainsi que les volumes associés. Cette analyse peut aider à comprendre la force et la signification des mouvements de prix basés sur les gaps et les volumes.
