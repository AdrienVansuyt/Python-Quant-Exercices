# **Correlation Heatmap entre Cryptomonnaies**
## **Introduction:**

Dans cet exercice, nous allons créer une heatmap de corrélation entre plusieurs cryptomonnaies populaires et surtout à grosse marketcap. La corrélation entre les actifs est une mesure importante pour diversifier un portefeuille et minimiser les risques. Comprendre les corrélations peut aider à identifier les cryptomonnaies qui se comportent de manière similaire ou différente.

```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des symboles des cryptomonnaies avec les plus grandes capitalisations boursières
tickers = [
    'BTC-USD', 'ETH-USD', 'BNB-USD', 'USDT-USD', 'USDC-USD', 'XRP-USD',
    'DOGE-USD', 'ADA-USD', 'MATIC-USD', 'DOT-USD', 'SOL-USD', 'LTC-USD',
    'TRX-USD', 'SHIB-USD', 'AVAX-USD', 'LINK-USD', 'BCH-USD'
]

# Télécharger les données de prix
data = yf.download(tickers, period='1y', interval='1d')

# Vérifier les cryptomonnaies pour lesquelles les données sont disponibles
available_tickers = [ticker for ticker in tickers if not data['Close'][ticker].isnull().all()]

# Extraire les prix de clôture et calculer les rendements quotidiens
close_prices = data['Close'][available_tickers]
returns = close_prices.pct_change().dropna()

# Calculer la matrice de corrélation
correlation_matrix = returns.corr()

# Créer la heatmap de corrélation
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Cryptocurrency Correlation Heatmap', fontsize=16)
plt.xlabel('Cryptocurrencies')
plt.ylabel('Cryptocurrencies')
plt.show()
```

# **Conclusion:**

Cette heatmap de corrélation montre les relations entre différentes cryptomonnaies avec les plus grandes capitalisations boursières. Des corrélations élevées indiquent que les cryptomonnaies se déplacent souvent ensemble, tandis que des corrélations faibles ou négatives suggèrent des mouvements indépendants ou opposés.
