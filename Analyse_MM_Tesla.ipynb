# **Analyse de la Moyenne Mobile pour les Actions de Tesla**
## **Introduction:**

Dans cet exercice, nous allons analyser les moyennes mobiles (SMA et EMA) pour les actions de Tesla (TSLA). Les moyennes mobiles sont des indicateurs techniques couramment utilisés pour lisser les fluctuations des prix et identifier les tendances directionnelles. Cela peut aider les traders à prendre des décisions éclairées sur les points d'entrée et de sortie.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Télécharger les données de Tesla
data = yf.download('TSLA', period='1y', interval='1d')

# Calculer les moyennes mobiles
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['EMA_20'] = data['Close'].ewm(span=20, adjust=False).mean()

# Tracer les prix de clôture et les moyennes mobiles
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['SMA_20'], label='20-day SMA', linestyle='--')
plt.plot(data['EMA_20'], label='20-day EMA', linestyle='--')
plt.title('Tesla Stock Price and Moving Averages', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## **Conclusion:**

En observant ce graphique, nous pouvons voir comment les moyennes mobiles simples (SMA) et exponentielles (EMA) suivent les prix de clôture de Tesla. Ces indicateurs peuvent être utilisés pour identifier les tendances et les points potentiels de retournement du marché, aidant ainsi à prendre des décisions de trading plus éclairées.
